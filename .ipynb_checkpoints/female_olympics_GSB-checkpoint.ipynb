```python
import pandas as pd
```

```python
# Load data
gold_female_sports_df = pd.read_csv('Female-Sports/female_olympics_gold.csv')
silver_female_sports_df = pd.read_csv('Female-Sports/female_olympics_silver.csv')
bronze_female_sports_df = pd.read_csv('Female-Sports/female_olympics_bronze.csv')
```

```python
# Group by Event - Gold
gold_df_by_event = gold_female_sports_df.groupby(['Event', 'Season', 'Sex', 'Medal'])

# Find median for Age, Height, Weight
gold_df_med_by_event = gold_df_by_event.median()
gold_df_med_by_event[["Age", "Height", "Weight"]]
```

```python
# Saving Dataframe into a csv file and json file
gold_df_med_by_event.to_csv('Female-Events/olympics_female_gold_bodycomp_age.csv', index=True)
gold_df_med_by_event.to_json('Female-Events/olympics_female_gold_bodycomp_age.json')
```

```python
# Group by Event - Silver
silver_df_by_event = silver_female_sports_df.groupby(['Event', 'Season', 'Sex', 'Medal'])

# Find median for Age, Height, Weight
silver_df_med_by_event = silver_df_by_event.median()
silver_df_med_by_event[["Age", "Height", "Weight"]]
```

```python
# Saving Dataframe into a csv file and json file
silver_df_med_by_event.to_csv('Female-Events/olympics_female_silver_bodycomp_age.csv', index=True)
silver_df_med_by_event.to_json('Female-Events/olympics_female_silver_bodycomp_age.json')
```

```python
# Group by Event - Bronze
bronze_df_by_event = bronze_female_sports_df.groupby(['Event', 'Season', 'Sex', 'Medal'])

# Find median for Age, Height, Weight
bronze_df_med_by_event = bronze_df_by_event.median()
bronze_df_med_by_event[["Age", "Height", "Weight"]]
```

```python
# Saving Dataframe into a csv file and json file
bronze_df_med_by_event.to_csv('Female-Events/olympics_female_bronze_bodycomp_age.csv', index=True)
bronze_df_med_by_event.to_json('Female-Events/olympics_female_bronze_bodycomp_age.json')
```
